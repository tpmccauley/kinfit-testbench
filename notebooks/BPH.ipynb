```python
hist_file = uproot.open('../files/his.root')
```

```python
tdr = hist_file['bphHistoSpecificDecay']
tdr.classes()
```

```python
hist, bins = tdr['massFull'].numpy()
width = 1.0*(bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2

plt.bar(center, hist, align='center', width=width, color='w', edgecolor='b')
plt.yscale('log')
plt.xlim(2,5)
plt.xlabel('')
plt.ylabel('')
plt.show()

```
